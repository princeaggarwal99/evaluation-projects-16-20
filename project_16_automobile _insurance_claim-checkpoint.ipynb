```python
# CREATING A FUNCTION FOR FINDING OUT THE BEST RANDOM STATE HAVING THE MAX r SCORE 
def maxr2_Score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=r_state,test_size=0.22)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2_score corr to random state',r_state,'is:',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2 score corr to ",final_r_state,'is',max_r_score)       
    return final_r_state 
```

```python
## tuning the parammeters
from sklearn.tree import DecisionTreeRegressor
rfr=RandomForestRegressor(n_estimators=1000)
knn=KNeighborsRegressor(n_neighbors=10)
svr=SVR(kernel='rbf')
ada=AdaBoostRegressor(learning_rate=0.1,n_estimators=200)
gbr=GradientBoostingRegressor(learning_rate=0.1,n_estimators=300)
dtr=DecisionTreeRegressor()
lr=LinearRegression()
```

```python
models = []
models.append(('randomforestREG', rfr))
models.append(('kneighborsREG', knn))
models.append(('svR', svr))
models.append(('ada boost REG', ada))
models.append(('gradient boosting REG', gbr))
models.append(('decisiontree REG', dtr))
models.append(('linear REG', lr))
```

```python
 #  as random forest in giving the best r2 score  of 92 with a ranom state of 97 
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=97,test_size=.22)
from sklearn.metrics import mean_squared_log_error,mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import cross_val_score
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
MSE=[]
MAE=[]
RMSE=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    R2=r2_score(y_test,pre)
    print("R2_score=",R2)
    score.append(R2*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='r2').mean()
    print('cross_val_sc=',sc)
    cvs.append(sc*100)
    print('\n')
    mse=mean_squared_error(y_test,pre)
    print('mse',mse)
    MSE.append(mse)
    print('\n')
    mae=mean_absolute_error(y_test,pre)
    print('mae',mae)
    MAE.append(mae)
    print('\n')
    Rmse=np.sqrt(mean_squared_error(y_test,pre))
    print('rmse',Rmse)
    RMSE.append(Rmse)
    print('\n\n')
    
```

```python
result=pd.DataFrame({'MODEL':Model,'R2 - SCORE':score,'CROSS_Val_Score':cvs,'MEAN SQ ERROR':MSE,'MEAN ABS ERROR':MAE,'ROOTMEAN SQ ERROR':RMSE})
result
```
