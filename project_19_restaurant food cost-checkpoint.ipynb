```python
# CREATING A FUNCTION FOR FINDING OUT THE BEST RANDOM STATE HAVING THE MAX R2 SCORE 
def maxr2_Score(regr,x,y):
    max_r_score=0
    for r_state in range(42,101):
        x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=r_state,test_size=0.33)
        regr.fit(x_train,y_train)
        y_pred=regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print('r2_score corr to random state',r_state,'is:',r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print()
    print()
    print("max r2 score corr to ",final_r_state,'is',max_r_score)       
    return final_r_state 
```

```python
# FINDING  BEST RANDOM STATE FOR LINEAR REG WITH HIGH R SCORE
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
r_state=maxr2_Score(lreg,x,y)
```

```python
## USING GRID SEARCH CV FOR FINDING THE BEST PARAMETERS FOR THE MODEL AND TUNING THE MODELS FOR BETTER PERFORMANCE
```

```python
# grid search CV = FINDING BEST PARAMETERS
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
```

```python
# BEST PARAMETERS FOR KNEIGHBORS REGRESSOR
parameters={'n_neighbors':[i+1 for i in range(10)]}
clf=GridSearchCV(knn,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
# BEST PARAMETERS FOR GRADIENT BOOSTING REGRESSOR
gbr= GradientBoostingRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000],'learning_rate':[0.001,0.01,0.1,1]}
clfg=GridSearchCV(gbr,parameters)
clfg.fit(x,y)
clfg.best_params_
```

```python
# BEST PARAMETERS FOR RANDOM FOREST REGRESSOR
rfr=RandomForestRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfr,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
## TUNING THE MODEL INTO THERI BEST PARAMETER FORM
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
rfr=RandomForestRegressor(n_estimators=1000)
gbr=GradientBoostingRegressor(n_estimators=1000)
knn=KNeighborsRegressor(n_neighbors=9)
svr=SVR(kernel='rbf')
ada=AdaBoostRegressor()
dtr=DecisionTreeRegressor()
lr=LinearRegression()
```

```python
## FINDING THE BEST R2 SCORE AND R STATE FOR GRADIENT BOOSTING REGRESSOR
r_state=maxr2_Score(gbr,x,y)
```

```python
## FINDING THE BEST R2 SCORE AND R STATE FOR RANDOM FOREST REGRESSOR
r_state=maxr2_Score(rfr,x,y)
```

```python
## best random state is occuring to be 64 having the best r2 score for model

x_train, x_test, y_train, y_test=train_test_split(x,y,random_state=64,test_size=0.33)
```

```python
models = []
models.append(('randomforestREG', rfr))
models.append(('kneighborsREG', knn))
models.append(('svR', svr))
models.append(('ada boost REG', ada))
models.append(('gradient boosting REG', gbr))
models.append(('decisiontree REG', dtr))
models.append(('linear REG', lr))
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
MSE=[]
MAE=[]
RMSE=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pre=model.predict(x_test)
    print('\n')
    R2=r2_score(y_test,pre)
    print("R2_score=",R2)
    score.append(R2*100)
    print('\n')
    sc=cross_val_score(model,x,y,cv=10,scoring='r2').mean()
    print('cross_val_sc=',sc)
    cvs.append(sc*100)
    print('\n')
    mse=mean_squared_error(y_test,pre)
    print('mse',mse)
    MSE.append(mse)
    print('\n')
    mae=mean_absolute_error(y_test,pre)
    print('mae',mae)
    MAE.append(mae)
    print('\n')
    Rmse=np.sqrt(mean_squared_error(y_test,pre))
    print('rmse',Rmse)
    RMSE.append(Rmse)
    print('\n\n')
    
```

```python
result=pd.DataFrame({'MODEL':Model,'R2 - SCORE':score,'CROSS_Val_Score':cvs,'MEAN SQ ERROR':MSE,'MEAN ABS ERROR':MAE,'ROOTMEAN SQ ERROR':RMSE})
result
```
