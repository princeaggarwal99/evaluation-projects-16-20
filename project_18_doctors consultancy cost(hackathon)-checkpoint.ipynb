```python
# IMPORTING  METRICS AND OTHER MODELS TO TRAIN AND TEST
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import  mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor,RandomForestRegressor
```

```python
# in hackacthon , results are evaluated by ROOT MEAN LOG SQUARE(RMLSE) METHODS 
```

```python
# defining a function to automate the process of log square in one go :-
def scoring(preds,y):
    preds = np.log1p(preds)
    y = np.log1p(y)
    return 1 - ((np.sum((preds-y)**2))/len(y))**1/2
```

```python
## linear regression model :-
lr=LinearRegression()
lr.fit(x_train,y_train)
ylreg=lr.predict(x_test)
print('SCORE',scoring(ylreg, y_test)*100)
print('MSE',mean_squared_error(ylreg, y_test))
print('MAE',mean_absolute_error(ylreg, y_test))
print('RMSE',np.sqrt(mean_squared_error(ylreg, y_test)))
```

```python
#USING GRID SEARCH FOR FINDING THE BEST PARAMETERS AND TUNE THE MODELS 
```

```python
# BEST PARAMETERS FOR KNEIGHBORS REGRESSOR
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
parameters={'n_neighbors':[i+1 for i in range(10)]}
clf=GridSearchCV(knn,parameters)
clf.fit(x,y)
clf.best_params_
```

```python
# BEST PARAMETERS FOR RANDOM FOREST REGRESSOR
rfr=RandomForestRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(rfr,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
# BEST PARAMETERS FOR GRADIENT BOOSTING REGRESSOR
gbr= GradientBoostingRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000],'learning_rate':[0.001,0.01,0.1,1]}
clfg=GridSearchCV(gbr,parameters)
clfg.fit(x,y)
clfg.best_params_
```

```python
# BEST PARAMETERS FOR RANDOM FOREST REGRESSOR
ada=AdaBoostRegressor()
parameters={'n_estimators':[1,10,50,100,200,300,400,500,1000]}
clfr=GridSearchCV(ada,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
# BEST PARAMETERS FOR RANDOM FOREST REGRESSOR
from sklearn.svm import SVR
svr=SVR()
parameters={'kernel':['rbf','poly','linear']}
clfr=GridSearchCV(svr,parameters)
clfr.fit(x,y)
clfr.best_params_
```

```python
## TUNING THE MODEL INTO THERI BEST PARAMETER FORM
from sklearn.tree import DecisionTreeRegressor
rfr=RandomForestRegressor(n_estimators=400)
gbr=GradientBoostingRegressor(n_estimators=300,learning_rate=0.1)
knn=KNeighborsRegressor(n_neighbors=10)
svr=SVR(kernel='rbf')
ada=AdaBoostRegressor(n_estimators=500)
dtr=DecisionTreeRegressor()
lr=LinearRegression()
```

```python
# now interpretting the best scoring model now ;
models = []
models.append(('randomforestREG', rfr))
models.append(('kneighborsREG', knn))
models.append(('svR', svr))
models.append(('ada boost REG', ada))
models.append(('gradient boosting REG', gbr))
models.append(('decisiontree REG', dtr))
models.append(('linear REG', lr))
```

```python
Model=[]
Scores=[]
cvs=[]
rocscore=[]
MSE=[]
MAE=[]
RMSE=[]
for name,model in models:
    print('********',name,'*******')
    print('\n')
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    preds=model.predict(x_test)
    print('\n')
    score=scoring(preds,y_test)
    print("score=",score)
    Scores.append(score*100)
    print('\n')
    mse=mean_squared_error(y_test,preds)
    print('mse',mse)
    MSE.append(mse)
    print('\n')
    mae=mean_absolute_error(y_test,preds)
    print('mae',mae)
    MAE.append(mae)
    print('\n')
    Rmse=np.sqrt(mean_squared_error(y_test,preds))
    print('rmse',Rmse)
    RMSE.append(Rmse)
    print('\n\n')
    
```

```python
result=pd.DataFrame({'MODEL':Model,'SCORE':Scores,'MEAN SQ ERROR':MSE,'MEAN ABS ERROR':MAE,'ROOTMEAN SQ ERROR':RMSE})
result
```
